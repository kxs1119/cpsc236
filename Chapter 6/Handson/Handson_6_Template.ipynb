## 12.1 Champion Counter
Create a program that reads a text file that contains a list of FIFA World Cup champions and determines the country that has won the most championships.

### Console:
```powershell

=======	        ====	=====	
Argentina       2       1978, 1986
Brazil	        5       1958, 1962, 1970, 1994, 2002
England	        1       1966		
France	        1       1998		
Germany	        4       1954, 1974, 1990, 2014
Italy	        4       1934, 1938, 1982, 2006
Spain	        1       2010
Uruguay	        2       1930, 1950
```

### Specifications:
- Use the attached text file named `world_cup_champions.txt` that contains data like this: 
```powershell
Year,Country,Coach,Captain 
1930,Uruguay,Alberto Suppici,JosÃ© Nasazzi 
1934,Italy,Vittorio Pozzo,Gianpiero Combi 
1938,Italy,Vittorio Pozzo,Giuseppe Meazza
...
```
- When the program starts, it should read the text file and use a dictionary to store the required data using the name of each country that has won the World Cup as the key.
- The program should compile the data shown above and display the countries alphabetically.



````python
import csv

def extractdata(file):
    # extracting the data from the text file 
    data = [] # creating an empty list 
    with open(file) as f:
        reader = csv.reader(f)
        for row in reader:
            data.append(row)
        return data

def splitdata(data):
    champions = {}
    for row in data:
        year = row[0]
        country = row[1]
        coach = row[2]
        captain = row[3]
        if country in champions:
            champions[country][0].append(year)
            champions[country][1] += 1
        else:
            champions[country] = [[year], 1]
    return champions

def printData(champions):
    # Print header
    print("Country".ljust(12), "Wins".ljust(7), "Years".ljust(10))
    print("="*30)
    
    # Print data
    for country, (years, wins) in sorted(champions.items()):
        print(country.ljust(12), str(wins).ljust(7), ', '.join(map(str, years)).ljust(10))
        
    



def main():
    data = extractdata('world_cup_champions.txt')
    champions = splitdata(data)
    results = printData(champions)
    print(results)
    

if __name__ == '__main__':
    main()
````
