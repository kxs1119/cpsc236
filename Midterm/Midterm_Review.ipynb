### Specifications
- The user should be able to check their balance and withdraw money.
- The user should have an initial balance of $1000
- Use a loop to allow the user to continue performing operations until they exit
- Validate user inputs and use functions to organize the code

````python
def check_balance(balance ):
    print("Balance $", balance)
    
def withdraw(amount, balance):
    if amount > balance[0]: 
        print("Insufficent Amounts")
        return
    elif amount < 0:
        print("Invalid Withdraw Amount")
        return 
    else:
        balance[0] = balance[0] - amount 
    return balance
def main():
    balance = [1000]
    option = 1 
    while True: 
        print("Choose an option: ")
        print("1.Check Balance")
        print("2.Withdraw Money")
        print("3. Exit")
        print()
        option = int(input("Enter Choice: "))
        
        if option == 1:
            print()
            check_balance(balance)
        
        elif option == 2:
            amount = input("Enter the withdraw amount:" )
            withdraw(amount, balance)
            
        elif option == 3:
            print("Goodbye")
            break
        
        else: 
            print("Invalid Statement. Please try again")

if __name__ == "__main__":
    main()
````

````python
import currency_converter as cc


def main():
    option = 'y'
    
    while option.lower() == 'y':
        print("Currency Converter")
        print()
        amount = int(input("Enter an amount in USD: "))
        eur_amount = cc.usd_to_eur(amount)
        print("$"+str(amount) + "in EUR: " + str(eur_amount))
        print()
        
        option = input("Continue(y/n)")
        
        if option == 'n':
              print("Goodbye!")

#if __name__ == "__main__":
main()
````
