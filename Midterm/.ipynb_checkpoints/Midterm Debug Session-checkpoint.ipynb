```python
def list_(movies):
    counter = 1
    for i in range(len(movies)):
        if len(movies) == 0:
            print("There are no movies in the list")
        else:
            print(counter, movies[i])
        counter += 1

def add(movies):
    if len(movies) == 100:
        print("Full Capacity")
        pass
    else:
        name = input("Name: ")
        year = int(input("Year: "))
        movies.append(name + ' ' + str(year))

def delete(movies):
    number = int(input("Number: "))
    movies.pop(number)
 
    

def main():
# Defing a list
    movies = ['Lilo and Stich 1988', 'Demon Slayer Movie 2023', 'Snow White 1943']

    print("COMMAND MENU")

    while True:
     # Printing command List
        print("list- list all movies")
        print("add- Add all movies")
        print("del- Delete all movies")
        print("exit- Exit all movies")

        option = input("Command: ")

      # Conditional for choices
        if option == "list":
            list_(movies)
        elif option == "add":
            add(movies)
        elif option == "del":
            delete(movies)
        elif option == "exit":
            print("Goodbye")
            break
        else:
            print(" Error! Please enter a valid choice")

# Calling the main function
if __name__ == '__main__':
    main()
```
