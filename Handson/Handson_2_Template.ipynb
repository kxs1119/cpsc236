## 3.1 Letter Grade Converter
Create a program that converts number grades to letter grades.

### Console:
```powershell
Letter Grade Converter

Enter numerical grade: 90 
Letter grade: A

Continue? (y/n): y

Enter numerical grade: 88 
Letter grade: A

Continue? (y/n): y

Enter numerical grade: 80 
Letter grade: B

Continue? (y/n): y

Enter numerical grade: 67 
Letter grade: C

Continue? (y/n): y

Enter numerical grade: 59 
Letter grade: F

Continue? (y/n): n

Bye!
```

### Specifications:
- Assume that the user will enter valid integers for the grades.
- The program should continue only if the user enters “y” or “Y” to continue.
- The grading criteria is as follows:

| Letter |	Range |
|---|-------|
| A |	88-100 |
| B |	80-87 |
| C |	67-79 |
| D |	60-66 |
| F |	<60 |

````python
print("Letter Grade Converter")
print()
# Get the data 
Choice = 'y' # Loop Counter
while Choice.lower() == 'y': 
    numberGrade = int(input("Enter numerical grade: "))
    if numberGrade >= 88 and numberGrade <= 100:
        print("Letter Grade: A")
    elif numberGrade >= 80 and numberGrade <=87:
        print("Letter Grade: B")
    elif numberGrade >= 67 and numberGrade <= 78:
        print("Letter Grade: C")
    elif numberGrade >= 60 and numberGrade <= 66:
        print("Letter Grade: D")
    else:
        print("Letter Grade: F")
    # Give the choice to restart
    print()
    userChoice = input("Continue? (y/n): ")
    print()
    if userChoice != Choice:
        print("Bye!")
        break
````

### Specifications:
- The program should calculate and display the cost of tipping at 15%, 20%, or 25%.
- Assume the user will enter valid data.
- The program should round results to a maximum of two decimal places.



````python
# Cost Input and Calculation 
mealCost = int(input("Enter the cost of the meal: "))
print()
smallTip =round( mealCost * float(.15),3)
mediumTip = round(mealCost + float(.20),3)
bigTip = round(mealCost + float(.25),3)

# Print out the results 
print("Tip Calculator")
print()

print("Cost of meal:", str(mealCost))
print()

# 15 tip
print("Tip amount:", str(smallTip))
print("Total amount:",str(smallTip + mealCost))
print()

# 20 tip
print("Tip amount:", str(mediumTip))
print("Total amount:",str(mediumTip + mealCost))
print()

#25 tip
print("Tip amount:", str(bigTip))
print("Total amount:",str(bigTip + mealCost))
print()
````

## 3.3 - Change Calculator
Create a program that calculates the coins needed to make change for the specified number of cents.

### Console
```powershell
Change Calculator

Enter number of cents (0-99): 99

Quarters: 3
Dimes:	2
Nickels: 0
Pennies: 4

Continue? (y/n): y

Enter number of cents (0-99): 55

Quarters: 2
Dimes:	0
Nickels: 1
Pennies: 0

Continue? (y/n): n 

Bye!
```

### Specifications
- The program should display the minimum number of quarters, dimes, nickels, and pennies that one needs to make up the specified number of cents.
- Assume that the user will enter a valid integer for the number of cents.
- The program should continue only if the user enters “y” or “Y” to continue.


````python
print("Change Calculator")
print()
Choice = 'y'
while Choice.lower() == 'y':
    #Variables 
    quarter = 25
    dime = 10
    nickel = 5
    penny = 1

    # Get data from user
    centCount = int(input("Enter number of cents(0-99): "))
    print()
    
    # Calculate data 
    quarterAmount = centCount // quarter 
    centCount %= quarter

    dimeAmount = centCount // dime
    centCount %= dime

    nickelAmount = centCount // nickel
    centCount %= nickel

    pennyAmount = centCount // penny 
    centCount %= penny

    #Display the data
    print("Quarter: ", quarterAmount)
    print("Dime :", dimeAmount)
    print("Nickel: ", nickelAmount)
    print("Pennies: ", pennyAmount)
    
    # print choice 
    print()
    userChoice = input("Continue (y/n): ")
    if userChoice != Choice:
        print("Bye!")
        break
````

## 3.4 - Table of Powers
Create a program that displays a table of squares and cubes for the specified range of numbers.

### Console
```powershell
Table of Powers

Start number: 90
Stop number: 100

Number	Squared	Cubed
======	=======	=====
90      8100	729000
91      8281	753571
92      8464	778688
93      8649	804357
94      8836	830584
95      9025	857375
96      9216	884736
97      9409	912673
98      9604	941192
99      9801	970299
100     10000	1000000
```

### Specifications
- The formulas for calculating squares and cubes are:

    `square = x ** 2`
    
    `cube = x ** 3`

- Use tabs to align the columns.
- Assume that the user will enter valid integers.
- Make sure the user enters a start integer that’s less than the stop integer. If the user enters a start integer that’s greater than the stop integer, display an error message and give the user a chance to enter the integers again.


````python
print("Table of Powers\n")

start = int(input("Start number: "))
stop = int(input("Stop number: "))
print()

print("Number\tSquared\tCubed")
print("======\t=======\t=====")
for x in range(start, stop + 1):
    squareNumbers = x ** 2
    cubeNumbers = x ** 3
    print(f"{x}\t{squareNumbers}\t{cubeNumbers}")
````
